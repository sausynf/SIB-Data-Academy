## Stacked Bar Plot

```python
dfgc = df.groupby(['PaymentMethod','Churn'])['MonthlyCharges'].sum().reset_index()
dfgc
```

```python
dfgc_piv = dfgc.pivot(index='PaymentMethod', columns='Churn', values='MonthlyCharges')
dfgc_piv
```

```python
dfgc_piv.plot(kind ='barh', stacked = True, figsize=(10,4))
```
